2. what is thepercentage of people who servived?

3. How many of each sex survived?

```python
survived_filter =  filename['Survived']==1
```

```python
filename[survived_filter]
```

```python
filename.groupby('Sex')['Survived'].sum()
```

4. What is the percentage of people that survived who paid a fare less than 10?

```python
fare_filter = filename['Fare'] < 10
filename.loc[fare_filter].groupby('Survived')['Fare'].mean()
```

```python
pps_df = filename[fare_filter]
pps_df['Survived'].mean()
```

```python
filename.loc[survived_filter & fare_filter].head(10)
```

5. what is the avarage age of those who didnt survive?

```python
filename.loc[~survived_filter, 'Age'].mean()
```

6. What is the avarage of those who did survive?

```python
filename.loc[survived_filter, 'Age'].mean()
```

7. What is the avarage age of those who did and didn't survived grouped by sex?

```python
filename.groupby(['Sex', 'Survived'])[['Age']].mean()
```
