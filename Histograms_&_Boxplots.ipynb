```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = '/content/catsvdogs.xlsx'
df = pd.read_excel(filename)
df.head()
```



* Plot a histogram of the Percetage of Dog Owners



```python
axes = sns.histplot(data=df, x='Percentage of Dog Owners')
axes.set(title="Dogs")
mean_rate = round(df['Percentage of Dog Owners'].mean(),2)
axes.axvline(mean_rate, color='k', ls="--",
           label=f"Percentage of Dog Owners={mean_rate}");
axes.legend();
```



* Plot a histogram of the Percentage of Cat Owners



```python
axs = sns.histplot(data=df, x='Percentage of Cat Owners')
# Create labels
axes.set(title="Cats");

mean_ratee = round(df['Percentage of Cat Owners'].mean(),2)
axes.axvline(mean_rate, color='k', ls="--",
           label=f"Percentage of Cat Owners={mean_rate}");
axes.legend();
```



* Create two boxplots on the same plot: one of the mean number of cats per household and the other of the mean number of dogs. An example of what this might look like 

```python
white_lines = dict(color='white')

plt.figure(figsize = (12, 5))
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8,5))
boxplots = axes.boxplot([df['Mean Number of Cats'],df['Mean Number of Dogs per household']],
           notch = True,
           labels=['Cats', 'Dogs'],
           widths = .7,
           patch_artist=True, 
           medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
           boxprops = dict(linestyle='--', linewidth=2, color='white', facecolor = 'blue', alpha = .4),
           whiskerprops=white_lines,
           flierprops=white_lines,
           capprops=white_lines
          );
# The more you understand any library, the more you can do
boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('orange')
plt.ylabel('Mean Number of Animals per State', fontsize = 20);
plt.xticks(fontsize = 16);
plt.yticks(fontsize = 16);
```
